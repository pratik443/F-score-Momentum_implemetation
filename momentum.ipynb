```python
def momentumProfit(formation,holding_period=1):
  begin_mes = formation-MonthEnd(12)
  end_mes = formation-MonthEnd(1)
  prices_end = prices.loc[end_mes]
  prices_begin = prices.loc[begin_mes]
  ret12 = prices_end/prices_begin-1
  ret12 = ret12.reset_index().dropna()
  ret12['quantiles'] = pd.qcut(ret12.iloc[:,1],10,labels=False)
  winners = ret12[ret12.quantiles==9]
  winnerret = prices.loc[formation+MonthEnd(holding_period),winners['index']]/prices.loc[formation,winners['index']]

  return np.mean(winnerret)
```

```python
#backtesting the strategy as given in papers but only for J=12
profits,dates,holding =[],[],[]
for i in (1,3,6):
  for j in prices.index[12:-i]: # we have to start from 12 months of starting as we want past 12 months
    profits.append(momentumProfit(j,i))
    dates.append(j+MonthEnd(i))
    holding.append(i)
```

```python
frame = pd.DataFrame({'MomentumProfits':profits,'HoldingPeriods':holding},index=dates)
```
