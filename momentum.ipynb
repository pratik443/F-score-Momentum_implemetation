```python
import yfinance as yf
import pandas as pd
import sqlalchemy
import datetime as dt
from pandas.tseries.offsets import MonthEnd
import numpy as np
from scipy.stats import percentileofscore as score
from statistics import mean
from math import floor
```

```python
tickersNIFTY = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[1].Symbol.to_list()
```

```python
tickersNIFTY
```

```python
data = []

for symb in tickersNIFTY:
  data.append(yf.download(f'{symb}.BO').reset_index())
```

```python
engine = sqlalchemy.create_engine('sqlite:///India.db')
```

```python
for frame,symbol in zip(data,tickersNIFTY):
  frame.to_sql(symbol, engine, index=False)
```

```python
df = pd.DataFrame()
```

```python
for name in tickersNIFTY:
  df = df.append(pd.read_sql(f'SELECT Date, "Adj Close" AS "{name}" FROM "{name}" WHERE Date > "2010-01-01"',engine))
```

```python
df = df.groupby('Date').sum()
```

```python
df.index = pd.to_datetime(df.index)
```

```python
df.index = pd.to_datetime(df.index)
```

```python
prices = df.resample('M').last()
prices
```

```python
mcg = []
for i in prices.columns:
    mcg.append(prices[i].pct_change().sum())
    
prices_momentum = pd.DataFrame(columns = ['symbol', 'mon_change'])
prices_momentum['symbol'] = prices.columns
prices_momentum['mon_change'] = mcg

# CALCULATING MOMENTUM

prices_momentum['momentum'] = 'N/A'
for i in range(len(prices_momentum)):
    prices_momentum.loc[i, 'momentum'] = score(prices_momentum.mon_change, prices_momentum.loc[i, 'mon_change'])/100
    
prices_momentum['momentum'] = prices_momentum['momentum'].astype(float)    
prices_momentum.head()
```

```python
top_picks = prices_momentum.nlargest(10, 'momentum')['symbol'].reset_index().drop('index', axis = 1)
top_picks
```
